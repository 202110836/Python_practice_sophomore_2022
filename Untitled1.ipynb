```python
# 1번 확인(어린이, 노인) -> 전체적으로 correct

def age_classify(age):
    cla = ''
    if age >= 60:
        cla = 'Elderly'
    elif age >= 40:
        cla = 'Older'
    elif age >= 30:
        cla = 'Adult'
    elif age >= 20:
        cla = 'Young Adult'
    elif age >= 10:
        cla = 'Teenager'
    elif age >= 5:
        cla = 'Kid'
    elif age > -1:
        cla = 'Baby'
    else:
        cla = 'Unknown'
    return cla
plt.figure(figsize = (10, 6))
file_df['Age_cla'] = file_df['Age'].apply(lambda x:age_classify(x))
orderr = ['Unknown', 'Baby', 'Kid', 'Teenager', 'Young Adult', 'Adult', 'Older', 'Elderly']
sns.barplot('Age_cla', 'Survived', hue = 'Sex', data = file_df, order = orderr)
```

```python
# 영향을 미치는 피처: Age, Pclass, Cabin(?), Sex
# 문자열: Cabin, Sex, 'Embarked' -> 인코딩

from sklearn import preprocessing

def encoding(file_df):
    features = ['Cabin', 'Sex', 'Embarked']
    for i in features:
        le = preprocessing.LabelEncoder()
        le = le.fit(file_df[i])
        file_df[i] = le.transform(file_df[i])
    return file_df

file_df = encoding(file_df)   #one hot??
file_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Null 처리 함수
def fillna(df):
    df['Age'].fillna(df['Age'].mean(),inplace=True)
    df['Cabin'].fillna('N',inplace=True)
    df['Embarked'].fillna('N',inplace=True)
    df['Fare'].fillna(0,inplace=True)
    return df

# 머신러닝 알고리즘에 불필요한 속성 제거
def drop_features(df):
    df.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
    return df

# 레이블 인코딩 수행. 
def format_features(df):
    df['Cabin'] = df['Cabin'].str[:1]
    features = ['Cabin','Sex','Embarked']
    for feature in features:
        le = LabelEncoder()
        le = le.fit(df[feature])
        df[feature] = le.transform(df[feature])
    return df

# 앞에서 설정한 Data Preprocessing 함수 호출
def transform_features(df):
    df = fillna(df)
    df = drop_features(df)
    df = format_features(df)
    return df
```

```python
from sklearn.model_selection import train_test_split
file_df = pd.read_csv('titanic_train.csv')
y_file_df = file_df['Survived']
X_file_df = file_df.drop('Survived', axis = 1)
X_file_df = transform_features(X_file_df)
X_train, X_test, y_train, y_test = train_test_split(X_file_df, y_file_df, test_size = 0.2, random_state =11)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 결정트리, Random Forest, 로지스틱 회귀를 위한 사이킷런 Classifier 클래스 생성
dt_clf = DecisionTreeClassifier(random_state=11)
rf_clf = RandomForestClassifier(random_state=11)
lr_clf = LogisticRegression()

# DecisionTreeClassifier 학습/예측/평가
dt_clf.fit(X_train , y_train)
dt_pred = dt_clf.predict(X_test)
print('DecisionTreeClassifier 정확도: {0:.4f}'.format(accuracy_score(y_test, dt_pred)))

# RandomForestClassifier 학습/예측/평가
rf_clf.fit(X_train , y_train)
rf_pred = rf_clf.predict(X_test)
print('RandomForestClassifier 정확도:{0:.4f}'.format(accuracy_score(y_test, rf_pred)))

# LogisticRegression 학습/예측/평가
lr_clf.fit(X_train , y_train)
lr_pred = lr_clf.predict(X_test)
print('LogisticRegression 정확도: {0:.4f}'.format(accuracy_score(y_test, lr_pred)))
```

```python
#교차검증(KFold) 사용 but 좋은 방법은 x
from sklearn.model_selection import KFold

def exec_kfold(clf, folds = 5):
    kfold = KFold(folds)
    scores = []
    
    for iter_count, (train_index, test_index) in enumerate(kfold.split(X_file_df)):
        X_train, X_test = X_file_df.values[train_index], X_file_df.values[test_index]
        y_train, y_test = y_file_df.values[train_index], y_file_df.values[test_index]
        
        clf.fit(X_train, y_train)
        clf_pred = clf.predict(X_test)
        accuracy = accuracy_score(y_test, clf_pred)
        scores.append(accuracy)
        print('교차 검증 {0} 정확도: {1:.4f}'.format(iter_count, accuracy))
    mean_score = np.mean(scores)
    print('평균 정확도: {0:.4f}'.format(mean_score))

exec_kfold(dt_clf, folds = 5)
```

```python
#교차검증(cross_val_score) 사용 보통 정도의 실용성
from sklearn.model_selection import cross_val_score

scores = cross_val_score(dt_clf, X_file_df.values, y_file_df.values, cv = 5)
print('교차검증별 정확도:', np.round(scores, 4))
print('평균 검증별 정확도:', np.round(np.mean(scores), 4))
```

```python
#교차검증(GridSearchCV) 가장 좋은 교차검증법
from sklearn.model_selection import GridSearchCV

para = {'max_depth':[2, 3, 5, 10], 'min_samples_split':[2, 3, 5], 'min_samples_leaf':[1, 5, 8]}
grid_dclf = GridSearchCV(dt_clf, para, 'accuracy', cv = 5)
grid_dclf.fit(X_train, y_train)
print('GridSearchCV 최적 하이퍼 파라미터 :',grid_dclf.best_params_)
print('GridSearchCV 최고 정확도: {0:.4f}'.format(grid_dclf.best_score_))
best_dclf = grid_dclf.best_estimator_

# GridSearchCV의 최적 하이퍼 파라미터로 학습된 Estimator로 예측 및 평가 수행. 
dpredictions = best_dclf.predict(X_test)
accuracy = accuracy_score(y_test , dpredictions)
print('테스트 세트에서의 DecisionTreeClassifier 정확도 : {0:.4f}'.format(accuracy))
```
